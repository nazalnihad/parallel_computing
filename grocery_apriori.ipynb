```python
df = pd.read_csv('/content/Groceries_dataset.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.dropna()

df["invoice"] = df['Member_number'].astype(str) + '_' + df["Date"].astype(str)
new_df = pd.pivot_table(df, index='invoice', columns='itemDescription', values='Member_number')

new_df = new_df +1
new_df = new_df.fillna(0)
new_df = new_df.applymap(lambda x: 1 if x!= 0 else 0)


apriori_df = apriori(new_df,
                     min_support=0.001,
                     use_colnames=True,
                     low_memory=True)
rules = association_rules(apriori_df,
                          metric='support',
                          min_threshold=0.001)

sorted_rules = rules.sort_values('lift', ascending=False)
sorted_rules

def calculate_support(df, item):
    total_transactions = df['invoice'].nunique()
    item_transactions = df[df['itemDescription'] == item]['invoice'].nunique()
    return item_transactions / total_transactions

print(calculate_support(df, 'whole milk'))
for threshold in [0.01, 0.005, 0.001]:
    frequent_itemsets = apriori(new_df, min_support=threshold, use_colnames=True)
    print(f"Min support {threshold}: {len(frequent_itemsets)} itemsets")

filtered_rules = sorted_rules[sorted_rules['support'] > 0.01]

sorted_filtered_rules = filtered_rules.sort_values('support', ascending=False)

for index, row in sorted_filtered_rules.iterrows():
    print(f"Rule: {row['antecedents']} -> {row['consequents']}")
    print(f"Support: {row['support']:.4f}")
    print(f"Confidence: {row['confidence']:.4f}")
    print(f"Lift: {row['lift']:.4f}")
    print("---")
```
